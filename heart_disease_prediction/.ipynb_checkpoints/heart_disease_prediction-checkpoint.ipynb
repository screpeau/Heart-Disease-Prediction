HEART DISEASE PREDICTION

    https://www.kaggle.com/ronitf/heart-disease-uci
    Original: http://archive.ics.uci.edu/ml/datasets/Heart+Disease

    We use the 13 fields as described below to predict the presence of heart disease in patients.

    n=303

ATTRIBUTES:

1. age (in years)
2. sex (0=female, 1=male)
3. cpt: chest pain type, (1=typical angina, 2=atypical angina, 3=non-anginal pain, 4=asympomatic)
4. restbp: resting blood pressure on admission to the hospital (in mm Hg)
5. chol: serum cholesterol (in mg/dl)
6. fbs: fasting blood sugar > 120 mg/dl (0=false, 1=true)
7. restecg: resting electrocardiographic results (integer 0-2)
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
8. thalach: maximum heart rate achieved (bpm)
9. exang: exercise induced angina (0=no, 1=yes)
10. oldpeak: ST depression induced by exercise relative to rest
11. slope: the slope of the peak exercise ST segment
12. ca: number of major vessels (integer 0-3) colored by fluoroscopy
13. thal: thalassemia (3=normal, 6=fixed defect, 7=reversible defect)
14. target: the presence of heart disease, integer from 0 (no presence) to 4

```python
# Must clean "?" values -> total of 6 missing values, 2 for 'ca' and 4 for 'thal'.
# Since these are categorical, we can impute with most common value

ca_mode = dataset.ca.mode()
thal_mode = dataset.thal.mode()

dataset.loc[dataset.ca == '?', 'ca'] = ca_mode
dataset.loc[dataset.thal == '?', 'thal'] = thal_mode
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = ['age','restbp','chol','thalach','oldpeak']
d_cols = ['sex','cp','fbs','restecg','exang','slope','ca','thal']

dataset_s = pd.get_dummies(dataset, columns = d_cols)
dataset_s[cols] = scaler.fit_transform(dataset_s[cols])
dataset_s.head()

# Why use dummy values over one hot encoding? Compare the two methods ***
```

```python
from sklearn.model_selection import train_test_split

y_train = dataset_s['target']
x_train = dataset_s.drop('target', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

forest = RandomForestClassifier(random_state=1)
forest.fit(x_train, y_train)
y_pred = forest.predict(x_test)
forest_scores = cross_val_score(forest, x_test, y_pred, cv=5)
print("Random Forest Classification Accuracy:", forest_scores.mean())
```
